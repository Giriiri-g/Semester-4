```python
A = [2, 7, 7, 11, 24, 24, 29, 36, 36]

def removeduplicates(arr):
  return list(set(arr))

def removeduplicates2(arr):
  B = []
  for i in arr:
    if i not in B:
      B.append(i)
  return B
removeduplicates2(A)
```

```python
A = [5, 4, 2, 8, 7, 1, 3, 9, 0, 6]

def splitArr(A):
  L, R = [], []
  pivot = A[0]
  for i in A:
    if i<pivot:
      L.append(i)
    else:
      R.append(i)
  return L, R

splitArr(A)
```

**Time Complexity: O(N)**
