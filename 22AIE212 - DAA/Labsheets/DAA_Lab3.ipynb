```python
def sumN(N):
  if N == 1:
    return 1
  else:
    return N + sumN(N-1)

sumN(5)
```

```python
def prodN(N):
  if N == 1:
    return 1
  else:
    return N * prodN(N-1)

prodN(5)
```

```python
def fibonacciN(N):
  if N == 2:
    return 1
  elif N==1:
    return 0
  return fibonacciN(N-1) + fibonacciN(N-2)

fibonacciN(5)
```

```python
def power(x, y):
  if y==0:
    return 1
  else:
    return x * power(x, y-1)

power(2, 10)
```

```python
def GCD(a, b):
  if b == 0:
    return a
  else:
    return GCD(b, a % b)

GCD(12, 16)
```

```python
def ArrayDisp(A):
  if A == []:
    return
  print(A[0], end=" ")
  ArrayDisp(A[1:])

ArrayDisp([5, 4, 3, 2, 1, 0])
```

```python
def revnum(N, r=0):
  if not N:
    return r
  else:
    return revnum(N//10, r*10 + N%10)


revnum(123001)
```

```python
def checkSort(A):
  if len(A) < 2:
    return True
  if A[0]>A[1]:
    return False
  else:
    return checkSort(A[1:])

checkSort([5, 4, 3, 2, 1]), checkSort([1, 2, 3, 4, 4])
```
