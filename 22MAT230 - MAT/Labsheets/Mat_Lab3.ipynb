```python
import numpy as np
from scipy.linalg import circulant, eig
```

```python
top_row = np.array([3, 5, 7])
circ_matrix = circulant(top_row)
circ_matrix
```

image.png

```python
def iscirculant(M):
  rows, cols = M.shape
  first_row = M[0]
  for i in range(1, rows):
    if not np.all(M[i] == np.roll(first_row, i)):
      return False
  return True

print(circ_matrix)
iscirculant(circ_matrix)
```

```python
C = np.array([[1, 2, 3],
              [3, 1, 2],
              [2, 3, 1]])

D = np.array([[5, 0, 4],
              [4, 5, 0],
              [0, 4, 5]])

CD = np.dot(C, D)
print(CD)
iscirculant(CD)
```
