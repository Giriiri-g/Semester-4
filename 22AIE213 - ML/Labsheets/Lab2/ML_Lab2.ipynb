```python
M1 = np.random.randint(0, 100, (10, 2))
disp(M1)
```

```python
Dist = []
for i in range(10):
  for j in range(10):
    Dist.append([i, j, EuclideanDist(M[:, 0:2][i], M1[j])])
Dist = np.array(Dist)
disp(Dist)
```

```python
from sklearn.metrics.pairwise import euclidean_distances
dist10x10 = euclidean_distances(M[:, 0:2], M1)
disp(dist10x10)
```

```python
print(f"{np.sum(dist10x10.reshape((100, 1)) == Dist[:, 2])}% of the Distances are the same")
```

```python
df = pd.read_csv("/content/diabetes_dataset.csv")
```

image.png

```python
df.head()
```

image.png

```python
X = df.drop("Outcome", axis=1)
X
```

```python
y = df["Outcome"]
y
```

image.png

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
print(X_test)
```

image.png

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler().fit(X_train)
```

```python
X_train =scaler.transform(X_train)
X_train
```

```python
X_test = scaler.transform(X_test)
X_test
```

image.png

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
y_pred
```

image.png

```python
from sklearn.metrics import classification_report, confusion_matrix
confus = confusion_matrix(y_test, y_pred)
print(confus)
print(classification_report(y_test, y_pred))
```

image.png

```python
confus[0][0] + confus[1][1]
```

image.png

```python
points = []
for i in range(2, 10):
  classifier = KNeighborsClassifier(n_neighbors=i)
  classifier.fit(X_train, y_train)
  y_pred = classifier.predict(X_test)
  confus = confusion_matrix(y_test, y_pred)
  points.append([i, confus[0][0] + confus[1][1]])
disp(np.array(points))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(np.array(points)[:, 0], np.array(points)[:, 1], 'r-')
```
